```python
sdf = pd.read_csv('spirals.speck', delim_whitespace=True,
                 names=['x','y','z','v1','v2','v3','v4','v5','v6','v7','v8'])
sdf
```

```python
#this file is available here : https://zenodo.org/record/2539456#.XZZU0edKh25
pdf = pd.read_csv('../../UniviewSNvTime/rawdata/plasticc_test_metadata.csv')
pdf
```

```python
yr = 2023
t = np.sort(Time(pdf['true_peakmjd'], format='mjd').decimalyear)
use = np.where(t > yr) #because there are some times in there back to 2001!)
t = t[use]
time = t[0:len(sdf)]
```

```python
#the Plastics time is binned, so I will need to randomize
tmin = min(time)
tmax = max(time)
print(tmin, tmax)
stime = np.random.random(len(time))*(tmax - tmin) + tmin
```

```python
shuffledf = sdf.sample(frac=1).reset_index(drop=True)
shuffledf['time'] = (np.sort(stime) - yr)*365.2425*24.
shuffledf
```

```python
outdf = pd.DataFrame()
outdf['x'] = shuffledf['x']
outdf['y'] = shuffledf['y']
outdf['z'] = shuffledf['z']
outdf['t'] = shuffledf['time']
outdf['f1'] = np.full(len(shuffledf),1)
outdf['t2'] = np.full(len(shuffledf),2)
outdf['f3'] = np.full(len(shuffledf),3)
outdf['f4'] = np.full(len(shuffledf),4)
outdf['f5'] = np.full(len(shuffledf),5)
outdf.to_csv("../data/SDSSLSSTSNe.raw",sep=' ', index=False, header=False)
```
